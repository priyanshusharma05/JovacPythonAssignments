PART 4: Students 

```python
data2 = pd.read_csv('StudentsPerformance.csv')
data2
```

```python
import matplotlib.pyplot as plt
```

```python
#1
avg_math= data2['math score'].mean()
avg_reading = data2['reading score'].mean()
avg_writing = data2['writing score'].mean()
print("Average Math Score:",avg_math)
print("Average Reading Score:",avg_reading)
print("Average Writing Score:",avg_writing)
```

```python
#2
data2.groupby('gender')['writing score'].mean().idxmax()
```

```python
#3
data2['total score'] = data2['math score'] + data2['reading score'] + data2['writing score']
data2['percentage'] = data2['total score'] / 3
data2
```

```python
#4
avg_scores_by_parent = data2.groupby('parental level of education')[['math score', 'reading score', 'writing score']].mean()

avg_scores_by_parent.plot(kind='bar')
plt.title('Average Scores by Parental Level of Education')
plt.ylabel('Average Score')
plt.show()
```

```python
#5
data2[(data2['math score'] > 90) & (data2['reading score'] > 90) & (data2['writing score'] > 90)]
```
